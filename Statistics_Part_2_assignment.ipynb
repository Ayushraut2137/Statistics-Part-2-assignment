1. What is hypothesis testing in statistics?
--> Hypothesis testing is a method used to decide whether there is enough evidence in a sample of data to infer that a certain condition is true for the entire population.

2. What is the null hypothesis, and how does it differ from the alternative hypothesis?
--> The null hypothesis is the default assumption that there is no effect, no difference, or no relationship.

The alternative hypothesis is what you‚Äôre trying to prove ‚Äî that there is an effect, a difference, or a relationship.

3. What is the significance level in hypothesis testing, and why is it important?
--> The significance level, often denoted as Œ± (alpha), is the threshold you set to decide whether a result is statistically significant.

It represents the probability of rejecting the null hypothesis (H‚ÇÄ) when it is actually true ‚Äî in other words, the chance of a false positive, also called a Type I error.

Sets the bar for evidence :

If your p-value ‚â§ Œ±, the result is considered statistically significant, and you reject the null hypothesis.

Controls the risk of being wrong:

By setting Œ±, you're deciding how much risk of error you're willing to accept.

Helps make consistent decisions:

It gives a clear rule for when to accept or reject H‚ÇÄ, reducing subjectivity.

4. What does a P-value represent in hypothesis testing?
--> The p-value is the probability of getting your observed results (or something more extreme) if the null hypothesis is true.

A small p-value means your data doesn‚Äôt look like what you'd expect under the null ‚Äî so you might have discovered something interesting.

5.  How do you interpret the P-value in hypothesis testing?
--> The p-value tells you how likely your sample results are if the null hypothesis (H‚ÇÄ) is true.

If p-value ‚â§ Œ± (e.g., 0.05):

 Reject the null hypothesis (H‚ÇÄ)

 Your results are statistically significant

 You have enough evidence to support the alternative hypothesis (H‚ÇÅ)

If p-value > Œ±:

 Fail to reject H‚ÇÄ

 Your results are not statistically significant

 There‚Äôs not enough evidence to support H‚ÇÅ (but it doesn‚Äôt mean H‚ÇÄ is true!)

6. What are Type 1 and Type 2 errors in hypothesis testing?
--> a. Type I Error (False Positive):

You reject the null hypothesis (H‚ÇÄ) when it's actually true.

The significance level (like 0.05) is the probability of making a Type I error.

b. Type II Error (False Negative):

You fail to reject H‚ÇÄ when it's actually false

The probability of making a Type II error (and 1 - Œ≤ is called the power of the test ‚Äî the ability to detect an actual effect).

19. What is the F-distribution, and when is it used in hypothesis testing?

20. What is an ANOVA test, and what are its assumptions?
--> ANOVA (Analysis of Variance) is a statistical test used to compare the means of three or more groups to determine if there is a statistically significant difference between them. It is an extension of the t-test, which compares the means of two groups, and it helps to avoid type I errors when comparing multiple groups.

ANOVA tests whether the variability between groups is significantly greater than the variability within groups. In other words, it checks if the differences in means between groups are larger than what would be expected by random chance.

 Assumptions of ANOVA:
For the results of an ANOVA test to be valid, certain assumptions must be met. These assumptions ensure that the test produces accurate and reliable results. The key assumptions are:

Independence of Observations:

The observations within each group must be independent of each other. This means that the value of one observation should not influence the value of another observation.

Normality:

The data within each group should follow a normal distribution. This assumption is particularly important when the sample sizes are small. If the sample sizes are large, the Central Limit Theorem helps in relaxing the normality assumption.

Homogeneity of Variances (Homoscedasticity):

The variance within each of the groups being compared should be approximately equal. This assumption is crucial because ANOVA compares the variability between and within groups. If the variances are unequal, the F-statistic might not be reliable.

A formal test for this assumption is the Levene's Test or Bartlett's Test, which test whether the variances are equal across groups.

21. What are the different types of ANOVA tests?
--> ANOVA (Analysis of Variance) has different types based on the structure of the experiment and the nature of the data being analyzed. The most common types of ANOVA are:

a. One-Way ANOVA

b. Two-Way ANOVA

c. Repeated Measures ANOVA

d. Multivariate Analysis of Variance (MANOVA)

a. One-Way ANOVA:

One-Way ANOVA is the simplest form of ANOVA. It is used when there is one independent variable (factor) with three or more levels (groups), and the goal is to compare the means of these different groups to see if there is a significant difference.

Use Case: Comparing the means of several groups to see if at least one group is different from the others.

Example: Testing the effectiveness of three different teaching methods (Method A, Method B, and Method C) on student performance.

b. Two-Way ANOVA:

Two-Way ANOVA is used when there are two independent variables (factors). This type of ANOVA helps to:

Assess the individual effects of each factor on the dependent variable.

Examine the interaction effect between the two factors.

Use Case: When you want to study the impact of two factors and their interaction on a dependent variable.

Example: Testing the impact of two factors, type of fertilizer (Factor 1: Fertilizer A, B, C) and watering frequency (Factor 2: Daily, Weekly), on plant growth.

c. Repeated Measures ANOVA:

Repeated Measures ANOVA is used when the same subjects are used in all groups or when multiple measurements are taken from the same group at different points in time. It is used to analyze the impact of an independent variable (or variables) on a dependent variable when the data are collected repeatedly from the same subjects.

Use Case: Testing how a treatment or condition affects the same group of subjects over time or under different conditions.

Example: Measuring the effect of a drug on blood pressure in the same group of patients at different time points (before treatment, during treatment, and after treatment).

d. Multivariate Analysis of Variance (MANOVA):

MANOVA is an extension of ANOVA that is used when there are multiple dependent variables that are correlated. It tests whether the means of the multiple dependent variables differ across groups.

Use Case: When you want to test the effect of one or more independent variables on two or more dependent variables simultaneously.

Example: Testing the effects of different diets on both weight loss and cholesterol levels.

22. What is the F-test, and how does it relate to hypothesis testing?
--> The F-test is a statistical test used to compare variances between two or more groups or populations. It is based on the F-distribution and is commonly used to test the hypothesis that two populations have the same variance. The F-test is a key part of Analysis of Variance (ANOVA) and can also be used in regression analysis to compare models.

The F-test is commonly used in the context of hypothesis testing to test whether there are significant differences between the variances of two or more groups. The basic steps for conducting an F-test in hypothesis testing are as follows:

State the Hypotheses:

Null Hypothesis (H‚ÇÄ): The variances of the populations (or groups) are equal. (For ANOVA, this would be that the group means are all equal.)

Alternative Hypothesis (H‚ÇÅ): The variances (or group means) are not equal.

Choose a Significance Level (Œ±):

Typically, ùõº = 0.05 (5% significance level), which means you are willing to accept a 5% chance of rejecting the null hypothesis when it is actually true.

Calculate the F-statistic:

Compute the variance between groups and the variance within groups and calculate the F-statistic.

Find the Critical Value:

Use an F-distribution table or statistical software to find the critical value of the F-statistic based on the degrees of freedom for the numerator and denominator, and the significance level (Œ±).


Make a Decision:

If the calculated F-statistic is greater than the critical value (or if the p-value is less than Œ±), reject the null hypothesis. This means there is a significant difference in variances (or group means).

If the calculated F-statistic is less than the critical value (or if the p-value is greater than Œ±), fail to reject the null hypothesis. This means the variances (or group means) are not significantly different.



















































































```python
# Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results

import numpy as np
from scipy import stats

sample_data = [50, 55, 53, 49, 52, 51, 56, 50, 52, 54]
sample_mean = np.mean(sample_data)
sample_size = len(sample_data)
sample_std_dev = np.std(sample_data, ddof=1)
population_mean = 50  # The known population mean
population_std_dev = 10  # The known population standard deviation (assumed)

# Perform the Z-test
z_score = (sample_mean - population_mean) / (population_std_dev / np.sqrt(sample_size))

# Calculate the p-value for a two-tailed test
p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

# Output the results
print("Sample Mean:", sample_mean)
print("Sample Size:", sample_size)
print("Sample Standard Deviation:", sample_std_dev)
print("Z-Score:", z_score)
print("P-Value:", p_value)

# Interpret the results
alpha = 0.05  # Significance level (5%)
if p_value < alpha:
    print("\nReject the null hypothesis: There is a significant difference between the sample mean and the population mean.")
else:
    print("\nFail to reject the null hypothesis: There is no significant difference between the sample mean and the population mean.")
```

```python
#  Perform a two-tailed Z-test using Python and visualize the decision region on a plot

import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Step 1: Simulate sample data
np.random.seed(42)  # For reproducibility
population_mean = 100
population_std_dev = 15
sample_size = 50

# Simulate sample data
sample_data = np.random.normal(loc=population_mean, scale=population_std_dev, size=sample_size)
sample_mean = np.mean(sample_data)
sample_std_dev = np.std(sample_data, ddof=1)  # Sample standard deviation
sample_size = len(sample_data)

# Step 2: Perform Z-test
z_score = (sample_mean - population_mean) / (population_std_dev / np.sqrt(sample_size))

# Two-tailed test - calculate the p-value
p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

# Step 3: Visualize the decision region
alpha = 0.05  # Significance level

# Calculate the critical values for two-tailed test
z_critical_left = stats.norm.ppf(alpha / 2)
z_critical_right = stats.norm.ppf(1 - alpha / 2)

# Create x values for plotting the normal distribution
x = np.linspace(population_mean - 4*population_std_dev, population_mean + 4*population_std_dev, 1000)
y = stats.norm.pdf(x, loc=population_mean, scale=population_std_dev)

# Plot the normal distribution
plt.plot(x, y, label='Normal Distribution', color='blue')

# Fill the rejection regions
plt.fill_between(x, y, 0, where=(x <= z_critical_left) | (x >= z_critical_right), color='red', alpha=0.5, label='Rejection Region')

# Add critical z-values and sample mean line
plt.axvline(z_critical_left, color='red', linestyle='dashed', label=f'Z-Critical Left = {z_critical_left:.2f}')
plt.axvline(z_critical_right, color='red', linestyle='dashed', label=f'Z-Critical Right = {z_critical_right:.2f}')
plt.axvline(sample_mean, color='green', linestyle='dashed', label=f'Sample Mean = {sample_mean:.2f}')

# Labels and title
plt.title('Two-Tailed Z-Test - Decision Region')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()
plt.grid(True)
plt.show()

# Step 4: Interpret the results
print("Sample Mean:", sample_mean)
print("Z-Score:", z_score)
print("P-Value:", p_value)

if p_value < alpha:
    print("\nReject the null hypothesis: There is a significant difference between the sample mean and the population mean.")
else:
    print("\nFail to reject the null hypothesis: There is no significant difference between the sample mean and the population mean.")

```

```python
#  Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing

import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

def visualize_type_errors(population_mean, population_std_dev, sample_size, alternative_mean, alpha=0.05):
    # Step 1: Generate the x values for plotting
    x = np.linspace(population_mean - 4 * population_std_dev, population_mean + 4 * population_std_dev, 1000)

    # Step 2: Create the normal distributions for the null hypothesis (H0) and alternative hypothesis (H1)
    # Null hypothesis (H0)
    y0 = stats.norm.pdf(x, loc=population_mean, scale=population_std_dev)
    # Alternative hypothesis (H1)
    y1 = stats.norm.pdf(x, loc=alternative_mean, scale=population_std_dev)

    # Step 3: Calculate critical values for the Z-test (two-tailed test)
    z_critical_left = stats.norm.ppf(alpha / 2)
    z_critical_right = stats.norm.ppf(1 - alpha / 2)

    # Step 4: Calculate the Type I error (Œ±) and Type II error (Œ≤)
    # Type I error: the area in the rejection region under the null hypothesis curve
    type_1_error = 1 - stats.norm.cdf(z_critical_right, loc=population_mean, scale=population_std_dev) + stats.norm.cdf(z_critical_left, loc=population_mean, scale=population_std_dev)

    # Type II error: the area in the non-rejection region under the alternative hypothesis curve
    type_2_error_left = stats.norm.cdf(z_critical_left, loc=alternative_mean, scale=population_std_dev)
    type_2_error_right = 1 - stats.norm.cdf(z_critical_right, loc=alternative_mean, scale=population_std_dev)
    type_2_error = type_2_error_left + type_2_error_right

    # Step 5: Visualize the decision region, Type I error, and Type II error
    plt.plot(x, y0, label='Null Hypothesis (H0)', color='blue')
    plt.plot(x, y1, label='Alternative Hypothesis (H1)', color='green')

    # Fill rejection regions (Type I Error regions)
    plt.fill_between(x, y0, 0, where=(x <= z_critical_left) | (x >= z_critical_right), color='red', alpha=0.5, label=f'Type I Error (Œ±) Area')

    # Fill Type II Error regions
    plt.fill_between(x, y1, 0, where=(x >= z_critical_left) & (x <= z_critical_right), color='yellow', alpha=0.5, label=f'Type II Error (Œ≤) Area')

    # Add critical value lines
    plt.axvline(z_critical_left, color='red', linestyle='dashed', label=f'Z-Critical Left = {z_critical_left:.2f}')
    plt.axvline(z_critical_right, color='red', linestyle='dashed', label=f'Z-Critical Right = {z_critical_right:.2f}')

    # Add labels and legend
    plt.title('Type I and Type II Errors Visualization')
    plt.xlabel('Value')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)
    plt.show()

    # Print error values
    print(f"Type I Error (Œ±): {type_1_error:.4f}")
    print(f"Type II Error (Œ≤): {type_2_error:.4f}")

# Example usage
population_mean = 100
population_std_dev = 15
sample_size = 30
alternative_mean = 105  # Hypothetical alternative population mean

visualize_type_errors(population_mean, population_std_dev, sample_size, alternative_mean, alpha=0.05)

```

```python
# Write a Python function to calculate the confidence interval for a sample mean and explain its significance

import numpy as np
from scipy import stats

def calculate_confidence_interval(sample_data, confidence_level=0.95, population_std=None):
    """
    Calculate the confidence interval for a sample mean.

    Parameters:
    sample_data (array-like): The sample data for which the confidence interval is calculated.
    confidence_level (float): The desired confidence level (default is 0.95).
    population_std (float or None): The population standard deviation. If None, sample standard deviation is used.

    Returns:
    tuple: Lower and upper bounds of the confidence interval.
    """
    # Step 1: Calculate sample mean and sample size
    sample_mean = np.mean(sample_data)
    sample_size = len(sample_data)

    # Step 2: Calculate sample standard deviation
    sample_std = np.std(sample_data, ddof=1)

    # Step 3: Determine whether to use Z-distribution (if population_std is provided) or T-distribution
    if population_std is not None:
        # Z-distribution (known population std)
        standard_error = population_std / np.sqrt(sample_size)
        z_score = stats.norm.ppf(1 - (1 - confidence_level) / 2)  # Z-score for two-tailed test
    else:
        # T-distribution (if population std is unknown)
        standard_error = sample_std / np.sqrt(sample_size)
        degrees_of_freedom = sample_size - 1
        t_score = stats.t.ppf(1 - (1 - confidence_level) / 2, df=degrees_of_freedom)  # T-score for two-tailed test

    # Step 4: Calculate the margin of error
    margin_of_error = t_score * standard_error if population_std is None else z_score * standard_error

    # Step 5: Calculate the confidence interval
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    return lower_bound, upper_bound

# Example usage
np.random.seed(42)  # For reproducibility
sample_data = np.random.normal(loc=50, scale=10, size=30)  # Simulating a sample with mean=50, std=10

confidence_interval = calculate_confidence_interval(sample_data, confidence_level=0.95)
print(f"The 95% confidence interval for the sample mean is: {confidence_interval}")
```

```python
#  Write a Python program to calculate the margin of error for a given confidence level using sample data

import numpy as np
from scipy import stats

def calculate_margin_of_error(sample_data, confidence_level=0.95, population_std=None):
    """
    Calculate the margin of error for a sample given a confidence level.

    Parameters:
    sample_data (array-like): The sample data.
    confidence_level (float): The desired confidence level (default is 0.95).
    population_std (float or None): The population standard deviation. If None, the sample standard deviation is used.

    Returns:
    float: The margin of error.
    """
    # Step 1: Calculate the sample mean and sample size
    sample_mean = np.mean(sample_data)
    sample_size = len(sample_data)

    # Step 2: Calculate the sample standard deviation
    sample_std = np.std(sample_data, ddof=1)

    # Step 3: Determine whether to use Z-distribution or T-distribution
    if population_std is not None:
        # Z-distribution (known population std)
        standard_error = population_std / np.sqrt(sample_size)
        z_score = stats.norm.ppf(1 - (1 - confidence_level) / 2)  # Z-score for two-tailed test
        margin_of_error = z_score * standard_error
    else:
        # T-distribution (unknown population std, use sample std)
        standard_error = sample_std / np.sqrt(sample_size)
        degrees_of_freedom = sample_size - 1
        t_score = stats.t.ppf(1 - (1 - confidence_level) / 2, df=degrees_of_freedom)  # T-score for two-tailed test
        margin_of_error = t_score * standard_error

    return margin_of_error

# Example usage
np.random.seed(42)  # For reproducibility
sample_data = np.random.normal(loc=50, scale=10, size=30)  # Simulating a sample with mean=50, std=10

# Calculate margin of error for 95% confidence level
margin_of_error = calculate_margin_of_error(sample_data, confidence_level=0.95)
print(f"The margin of error at 95% confidence level is: {margin_of_error:.4f}")
```

```python
# Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process

def bayesian_inference(prior, likelihood, evidence, false_positive_rate):
    """
    Perform Bayesian inference using Bayes' Theorem.

    Parameters:
    prior (float): Prior probability P(H), the initial belief about the hypothesis.
    likelihood (float): Likelihood P(E|H), the probability of observing the evidence given the hypothesis.
    evidence (float): The total probability of observing the evidence P(E).
    false_positive_rate (float): The false positive rate P(E|~H), the probability of observing the evidence assuming the hypothesis is false.

    Returns:
    float: The posterior probability P(H|E), the updated probability of the hypothesis given the evidence.
    """
    # Bayes' Theorem: P(H|E) = (P(E|H) * P(H)) / P(E)
    posterior = (likelihood * prior) / evidence
    return posterior

def calculate_evidence(likelihood, prior, false_positive_rate):

    # P(E) = P(E|H) * P(H) + P(E|~H) * P(~H)
    # P(~H) = 1 - P(H)
    p_not_h = 1 - prior
    evidence = (likelihood * prior) + (false_positive_rate * p_not_h)
    return evidence

# Given values:
prior = 0.01  # Prior probability of having the disease (1%)
likelihood = 0.99  # Probability of a positive test result given the person has the disease
false_positive_rate = 0.05  # Probability of a false positive (person doesn't have the disease, but the test is positive)

# Step 1: Calculate the evidence P(E)
evidence = calculate_evidence(likelihood, prior, false_positive_rate)

# Step 2: Apply Bayes' Theorem to calculate the posterior probability P(H|E)
posterior = bayesian_inference(prior, likelihood, evidence, false_positive_rate)

# Output the result
print(f"The posterior probability of having the disease given a positive test result is: {posterior:.4f}")
```

```python
# Write a Python program to calculate the expected frequencies for a Chi-square test based on observed data

import numpy as np

def calculate_expected_frequencies(observed):
    # Get the row totals and column totals
    row_totals = observed.sum(axis=1)
    column_totals = observed.sum(axis=0)

    # Calculate the grand total
    grand_total = observed.sum()

    # Calculate the expected frequencies using the formula
    expected = np.outer(row_totals, column_totals) / grand_total

    return expected

# Given observed data (contingency table)
observed_data = np.array([[30, 70],  # Male: [Smoker, Non-smoker]
                          [20, 80]]) # Female: [Smoker, Non-smoker]

# Calculate expected frequencies
expected_frequencies = calculate_expected_frequencies(observed_data)

# Print the observed and expected frequencies
print("Observed Frequencies:\n", observed_data)
print("\nExpected Frequencies:\n", expected_frequencies)
```

```python
#  Implement an F-test using Python to compare the variances of two random samples

import numpy as np
from scipy.stats import f

# Function to perform the F-test for comparing variances
def f_test(sample1, sample2, alpha=0.05):
    # Calculate the variances of the two samples
    var1 = np.var(sample1, ddof=1)  # sample variance (ddof=1 for sample variance)
    var2 = np.var(sample2, ddof=1)

    # Calculate the F-statistic (larger variance / smaller variance)
    if var1 > var2:
        F_statistic = var1 / var2
        df1 = len(sample1) - 1  # degrees of freedom for sample 1
        df2 = len(sample2) - 1  # degrees of freedom for sample 2
    else:
        F_statistic = var2 / var1
        df1 = len(sample2) - 1
        df2 = len(sample1) - 1

    # Calculate the p-value using the cumulative distribution function (CDF) of the F-distribution
    p_value = 1 - f.cdf(F_statistic, df1, df2)

    # Output the results
    print(f"F-statistic: {F_statistic:.4f}")
    print(f"P-value: {p_value:.4f}")

    # Interpretation of the results
    if p_value < alpha:
        print("\nWe reject the null hypothesis: The variances are significantly different.")
    else:
        print("\nWe fail to reject the null hypothesis: The variances are not significantly different.")

# Example: Simulate two random samples with different variances
np.random.seed(0)  # For reproducibility
sample1 = np.random.normal(loc=0, scale=1, size=50)  # Sample 1 with mean 0 and std 1
sample2 = np.random.normal(loc=0, scale=2, size=50)  # Sample 2 with mean 0 and std 2

# Perform the F-test
f_test(sample1, sample2)
```

```python
# Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns

# Function to check normality assumption using Shapiro-Wilk test
def check_normality(data):
    p_values = []
    for i, group in enumerate(data):
        stat, p_value = stats.shapiro(group)
        print(f"Shapiro-Wilk Test for Group {i+1}: p-value = {p_value:.4f}")
        p_values.append(p_value)
    return p_values

# Function to check the homogeneity of variances using Levene's Test
def check_equal_variance(*samples):
    stat, p_value = stats.levene(*samples)
    print(f"Levene's Test for Equal Variance: p-value = {p_value:.4f}")
    return p_value

# Function to plot the data for visual inspection of normality
def plot_normality(data):
    for i, group in enumerate(data):
        sns.distplot(group, label=f"Group {i+1}", kde=True, hist=True)
    plt.legend()
    plt.title("Distribution of Groups")
    plt.show()

# Function to create Q-Q plots to visually inspect normality
def plot_qq(data):
    for i, group in enumerate(data):
        stats.probplot(group, dist="norm", plot=plt)
        plt.title(f"Q-Q plot for Group {i+1}")
        plt.show()

# Function to perform all checks
def check_anova_assumptions(*samples):
    # Check Normality using Shapiro-Wilk test
    print("\nChecking Normality Assumption:")
    normality_p_values = check_normality(samples)

    # Plot distributions for normality check
    plot_normality(samples)

    # Check Homogeneity of Variances using Levene's test
    print("\nChecking Equal Variance Assumption:")
    p_value_variance = check_equal_variance(*samples)

    # Perform Q-Q plots for visual inspection
    plot_qq(samples)

    # Return results
    normality_check = all(p > 0.05 for p in normality_p_values)  # Normality assumption holds if all p-values > 0.05
    variance_check = p_value_variance > 0.05  # Homogeneity of variance holds if p-value > 0.05

    return normality_check, variance_check

# Example usage: Simulate data for three groups
np.random.seed(0)
group1 = np.random.normal(loc=5, scale=1, size=30)
group2 = np.random.normal(loc=5.5, scale=1, size=30)
group3 = np.random.normal(loc=6, scale=1, size=30)

# Check assumptions for ANOVA
normality_check, variance_check = check_anova_assumptions(group1, group2, group3)

# Final decision based on assumption checks
if normality_check and variance_check:
    print("\nAssumptions for ANOVA are met: Proceed with the ANOVA test.")
else:
    print("\nAssumptions for ANOVA are not met: Consider transformations or different tests.")
```

```python
# Perform a two-way ANOVA test using Python to study the interaction between two factors and visualize the results

import numpy as np
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols
import matplotlib.pyplot as plt
import seaborn as sns

# Simulating Data
np.random.seed(0)
factor1 = np.random.choice(['A', 'B'], size=60)  # Factor 1 with levels 'A' and 'B'
factor2 = np.random.choice(['X', 'Y', 'Z'], size=60)  # Factor 2 with levels 'X', 'Y', 'Z'
response = np.random.normal(loc=0, scale=1, size=60)  # Dependent variable

# Modify data to introduce some interaction
response[(factor1 == 'A') & (factor2 == 'X')] += 1
response[(factor1 == 'B') & (factor2 == 'Y')] -= 1

# Create a DataFrame
data = pd.DataFrame({'Factor1': factor1, 'Factor2': factor2, 'Response': response})

# Perform Two-Way ANOVA
model = ols('Response ~ Factor1 * Factor2', data=data).fit()  # 'Factor1 * Factor2' includes both main effects and interaction
anova_table = sm.stats.anova_lm(model, typ=2)  # Type II ANOVA table

# Display the ANOVA table
print(anova_table)

# Visualization of the interaction effect
plt.figure(figsize=(8, 6))
sns.pointplot(x='Factor1', y='Response', hue='Factor2', data=data, dodge=True, markers='o', linestyles='-', capsize=5)
plt.title('Interaction Plot for Two-Way ANOVA')
plt.ylabel('Response')
plt.xlabel('Factor 1')
plt.show()
```

```python
#  Perform a one-way ANOVA test in Python and visualize the results with boxplots to compare group means

import numpy as np
import pandas as pd
import scipy.stats as stats
import seaborn as sns
import matplotlib.pyplot as plt

# Simulate data for three different groups
np.random.seed(0)

# Simulating data for three groups with different means
group1 = np.random.normal(loc=20, scale=5, size=50)  # Group 1: mean=20, std=5
group2 = np.random.normal(loc=22, scale=5, size=50)  # Group 2: mean=22, std=5
group3 = np.random.normal(loc=25, scale=5, size=50)  # Group 3: mean=25, std=5

# Create a DataFrame to hold the data
data = pd.DataFrame({
    'Group': ['Group 1'] * 50 + ['Group 2'] * 50 + ['Group 3'] * 50,
    'Values': np.concatenate([group1, group2, group3])
})

# Perform the one-way ANOVA
f_statistic, p_value = stats.f_oneway(group1, group2, group3)

# Print the ANOVA results
print(f"F-statistic: {f_statistic}")
print(f"P-value: {p_value}")

# Interpretation of the results
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference between the group means.")
else:
    print("Fail to reject the null hypothesis: No significant difference between the group means.")

# Visualization of the results using boxplots
plt.figure(figsize=(8, 6))
sns.boxplot(x='Group', y='Values', data=data, palette="Set2")
plt.title('One-Way ANOVA: Comparison of Group Means')
plt.xlabel('Group')
plt.ylabel('Values')
plt.show()
```

```python
#  Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results

import numpy as np
import scipy.stats as stats

# Simulate data (sample data for a population)
np.random.seed(0)
sample_data = np.random.normal(loc=50, scale=10, size=100)  # mean=50, std=10, sample size=100

# Define the hypothesized population variance (sigma_0^2)
sigma_0_squared = 100  # hypothesized population variance (std^2 = 10^2 = 100)

# Calculate the sample variance (s^2)
sample_variance = np.var(sample_data, ddof=1)  # ddof=1 for sample variance

# Sample size (n)
n = len(sample_data)

# Calculate the Chi-square test statistic
chi_square_stat = (n - 1) * sample_variance / sigma_0_squared

# Degrees of freedom
df = n - 1

# Significance level
alpha = 0.05

# Calculate the critical values for the two-tailed test
chi_square_critical_lower = stats.chi2.ppf(alpha / 2, df)
chi_square_critical_upper = stats.chi2.ppf(1 - alpha / 2, df)

# Print the test statistic and critical values
print(f"Chi-square test statistic: {chi_square_stat}")
print(f"Chi-square critical values: Lower = {chi_square_critical_lower}, Upper = {chi_square_critical_upper}")

# Perform the hypothesis test
if chi_square_stat < chi_square_critical_lower or chi_square_stat > chi_square_critical_upper:
    print("Reject the null hypothesis: The population variance is significantly different from the hypothesized variance.")
else:
    print("Fail to reject the null hypothesis: The population variance is not significantly different from the hypothesized variance.")
```

```python
# D Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results

import numpy as np
import scipy.stats as stats

# Simulate a sample from a normal distribution
np.random.seed(42)
sample = np.random.normal(loc=100, scale=15, size=30)  # mean=100, std=15, n=30

# Hypothesized population variance
sigma_0_squared = 225  # (15^2)

# Sample variance
sample_variance = np.var(sample, ddof=1)
n = len(sample)
df = n - 1

# Chi-square test statistic
chi2_stat = (df * sample_variance) / sigma_0_squared

# Two-tailed test: significance level
alpha = 0.05
chi2_crit_lower = stats.chi2.ppf(alpha / 2, df)
chi2_crit_upper = stats.chi2.ppf(1 - alpha / 2, df)

# Print results
print(f"Sample Variance: {sample_variance:.2f}")
print(f"Chi-square statistic: {chi2_stat:.2f}")
print(f"Chi-square critical values: ({chi2_crit_lower:.2f}, {chi2_crit_upper:.2f})")

# Decision
if chi2_stat < chi2_crit_lower or chi2_stat > chi2_crit_upper:
    print("‚ùå Reject the null hypothesis: The population variance is significantly different.")
else:
    print("‚úÖ Fail to reject the null hypothesis: No significant difference in population variance.")
```

```python
#  Write a Python script to perform a Z-test for comparing proportions between two datasets or groups

import numpy as np
import scipy.stats as stats

def z_test_proportions(x1, n1, x2, n2, alpha=0.05):
    # Sample proportions
    p1 = x1 / n1
    p2 = x2 / n2

    # Pooled proportion
    p_pool = (x1 + x2) / (n1 + n2)

    # Z statistic
    z = (p1 - p2) / np.sqrt(p_pool * (1 - p_pool) * (1/n1 + 1/n2))

    # Two-tailed p-value
    p_value = 2 * (1 - stats.norm.cdf(abs(z)))

    # Critical z for 95% confidence
    z_crit = stats.norm.ppf(1 - alpha / 2)

    print(f"Proportion 1 (p1): {p1:.4f}")
    print(f"Proportion 2 (p2): {p2:.4f}")
    print(f"Z-statistic: {z:.4f}")
    print(f"P-value: {p_value:.4f}")
    print(f"Critical Z-value: ¬±{z_crit:.2f}")

    if abs(z) > z_crit:
        print("‚ùå Reject the null hypothesis: Proportions are significantly different.")
    else:
        print("‚úÖ Fail to reject the null hypothesis: No significant difference in proportions.")


z_test_proportions(x1=60, n1=100, x2=45, n2=100)
```

```python
# Implement an F-test for comparing the variances of two datasets, then interpret and visualize the results

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def f_test_variances(data1, data2, alpha=0.05):
    # Sample sizes
    n1, n2 = len(data1), len(data2)

    # Sample variances
    var1, var2 = np.var(data1, ddof=1), np.var(data2, ddof=1)

    # Make sure F >= 1 by dividing the larger variance by the smaller
    if var1 > var2:
        f_stat = var1 / var2
        df1, df2 = n1 - 1, n2 - 1
    else:
        f_stat = var2 / var1
        df1, df2 = n2 - 1, n1 - 1

    # Critical F-values for two-tailed test
    f_crit_low = stats.f.ppf(alpha / 2, df1, df2)
    f_crit_high = stats.f.ppf(1 - alpha / 2, df1, df2)

    # P-value
    p_value = 2 * min(stats.f.cdf(f_stat, df1, df2), 1 - stats.f.cdf(f_stat, df1, df2))

    # Print results
    print(f"Sample Variance 1: {var1:.4f}")
    print(f"Sample Variance 2: {var2:.4f}")
    print(f"F-statistic: {f_stat:.4f}")
    print(f"Critical F-values: ({f_crit_low:.4f}, {f_crit_high:.4f})")
    print(f"P-value: {p_value:.4f}")

    if f_stat < f_crit_low or f_stat > f_crit_high:
        print("‚ùå Reject H‚ÇÄ: The variances are significantly different.")
    else:
        print("‚úÖ Fail to reject H‚ÇÄ: No significant difference in variances.")

    # Visualization
    x = np.linspace(0.01, 5, 500)
    y = stats.f.pdf(x, df1, df2)

    plt.figure(figsize=(8, 4))
    plt.plot(x, y, label="F-distribution", color='blue')
    plt.axvline(f_stat, color='red', linestyle='--', label=f"F-statistic = {f_stat:.2f}")
    plt.axvline(f_crit_low, color='green', linestyle='--', label="Lower Critical Value")
    plt.axvline(f_crit_high, color='green', linestyle='--', label="Upper Critical Value")
    plt.title("F-Test: Comparison of Two Variances")
    plt.xlabel("F value")
    plt.ylabel("Probability Density")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

# Simulate two samples
np.random.seed(0)
data1 = np.random.normal(50, 10, 30)   # std = 10
data2 = np.random.normal(50, 15, 30)   # std = 15

# Run the F-test
f_test_variances(data1, data2)
```

```python
#  Perform a Chi-square test for goodness of fit with simulated data and analyze the results

import numpy as np
from scipy.stats import chisquare
import matplotlib.pyplot as plt

# Simulate observed data (e.g., rolling a 6-sided die 120 times)
np.random.seed(42)
observed = np.random.choice([1, 2, 3, 4, 5, 6], size=120)
observed_counts = np.bincount(observed)[1:]  # ignore 0th bin (dice starts from 1)

# Define expected frequencies assuming fair die (uniform distribution)
expected_counts = [120 / 6] * 6  # 20 expected for each side

# Chi-square test
chi_stat, p_value = chisquare(f_obs=observed_counts, f_exp=expected_counts)

# Display results
print("Observed Counts:", observed_counts)
print("Expected Counts:", expected_counts)
print(f"Chi-square Statistic: {chi_stat:.4f}")
print(f"P-value: {p_value:.4f}")

if p_value < 0.05:
    print("‚ùå Reject H‚ÇÄ: The data does not fit the expected distribution.")
else:
    print("‚úÖ Fail to reject H‚ÇÄ: The data fits the expected distribution.")

# Plotting
labels = ['1', '2', '3', '4', '5', '6']
x = np.arange(len(labels))
width = 0.35

fig, ax = plt.subplots()
bars1 = ax.bar(x - width/2, observed_counts, width, label='Observed')
bars2 = ax.bar(x + width/2, expected_counts, width, label='Expected', color='orange')

ax.set_xlabel('Die Face')
ax.set_ylabel('Frequency')
ax.set_title('Chi-square Goodness of Fit Test: Die Roll Frequencies')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()
plt.tight_layout()
plt.show()
```
